```python
df.fillna(method='ffill', inplace=True)
```

```python
sales_data = df[['Category','Sub-Category','Sales','Profit']]
sales_data.head()
```

```python
high_profit = df[df['Profit'] > 100]

high_profit.head()
```

```python
technology = df[df['Category'] == 'Technology']

technology.head()
```

```python
print("Duplicates:", df.duplicated().sum())

df.drop_duplicates(inplace=True)

print("Duplicates After Removal:", df.duplicated().sum())
```

```python
df['Profit_Margin'] = (df['Profit'] / df['Sales']) * 100

df[['Sales','Profit','Profit_Margin']].head()
```

```python
df['Revenue_Per_Unit'] = df['Sales'] / df['Quantity']
```

```python
category_sales = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)

print(category_sales)
```

```python
category_profit = df.groupby('Category')['Profit'].sum().sort_values(ascending=False)

print(category_profit)
```

```python
region_sales = df.groupby('Region')['Sales'].sum().sort_values(ascending=False)

print(region_sales)
```

```python
top_products = df.groupby('Product Name')['Sales'].sum()

top_products.sort_values(ascending=False).head(10)
```

```python
df.to_csv("cleaned_superstore.csv", index=False)

print("File Saved Successfully")
```
